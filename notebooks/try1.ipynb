```python
from sqlalchemy import create_engine
import pandas as pd
import urllib

# 1. Configuration
user = "daniel"
password = "db2dani"
host = "52.211.123.34"
port = "25010"
database = "IEMASTER"

# 2. Connection String (using the pyodbc method we discussed)
# It's safer to wrap the connection parameters to handle special characters
params = urllib.parse.quote_plus(
    f"DRIVER={{IBM DB2 ODBC DRIVER}};DATABASE={database};"
    f"HOSTNAME={host};PORT={port};PROTOCOL=TCPIP;"
    f"UID={user};PWD={password};"
)
connection_string = f"db2+pyodbc:///?odbc_connect={params}"

# 3. Initialize Engine
engine = create_engine(connection_string)

# 4. Query and Load Data
query = "SELECT * FROM IEPLANE.FLIGHTS FETCH FIRST 10 ROWS ONLY"

try:
    # Note the corrected syntax here: pd.read_sql(query, engine)
    df = pd.read_sql(query, engine)
    
    # Now .head() will work because df is a real DataFrame
    print(df.head())
    
except Exception as e:
    print(f"An error occurred: {e}")
```
